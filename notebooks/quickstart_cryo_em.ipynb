## Generate Density Maps
We'll simulate how a structure looks at different resolutions.

```python
import matplotlib.pyplot as plt
import numpy as np

from synth_cryo_em.generator import generate_map

# Create dummy coordinates (a simple triangle)
coords = np.array([[0, 0, 0], [5, 0, 0], [0, 5, 0]])

# Generate maps at 2A and 8A resolution
map_high = generate_map(coords, resolution=2.0, grid_spacing=0.5)
map_low = generate_map(coords, resolution=8.0, grid_spacing=0.5)

# Visualize a slice of the density
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(map_high[map_high.shape[0] // 2], cmap="Blues")
plt.title("High Res (2Å)")
plt.subplot(122)
plt.imshow(map_low[map_low.shape[0] // 2], cmap="Blues")
plt.title("Low Res (8Å)")
plt.show()
```
