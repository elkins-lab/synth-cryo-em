## Generate Density Maps
We'll simulate how a structure looks at different resolutions.

First, we'll create a minimal PDB file to work with.

```python
import matplotlib.pyplot as plt
import numpy as np

from synth_cryo_em.core import generate_density_map

pdb_content = """ATOM      1  N   ALA A   1       0.000   0.000   0.000  1.00  0.00           N
ATOM      2  CA  ALA A   1       1.458   0.000   0.000  1.00  0.00           C
ATOM      3  C   ALA A   1       2.009   1.362   0.000  1.00  0.00           C
ATOM      4  O   ALA A   1       1.282   2.257   0.000  1.00  0.00           O
END"""
with open("test.pdb", "w") as f:
    f.write(pdb_content)

# Generate maps at 2A and 8A resolution
# generate_density_map returns (grid, origin)
grid_high, _ = generate_density_map("test.pdb", resolution=2.0, grid_spacing=0.5)
grid_low, _ = generate_density_map("test.pdb", resolution=8.0, grid_spacing=0.5)

# Convert gemmi grid to numpy array
map_high = np.array(grid_high, copy=False)
map_low = np.array(grid_low, copy=False)

# Visualize a slice of the density
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(map_high[map_high.shape[0] // 2], cmap="Blues")
plt.title("High Res (2Å)")
plt.subplot(122)
plt.imshow(map_low[map_low.shape[0] // 2], cmap="Blues")
plt.title("Low Res (8Å)")
plt.show()
```
